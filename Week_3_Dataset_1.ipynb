# To check missing values

```python
data.isna().sum()
```

To check column in dataset

```python
data.columns = data.columns.str.lower().str.replace(' ','_')
```

```python
data.columns
```

# Replacing the missing values in the cloumn "Age"

```python
data.age
```

```python
avg_age = data.age.mean()
```

```python
data.age= data.age.fillna(avg_age)
data.age = data.age.round(decimals=1)
data.age
```

```python
data.age.isna().sum()
```

# Salary

```python
data['salary'] = data['salary'].apply(lambda x:x.replace('$',''))
```

```python
data.salary
```

```python
data['age'] = data['age'].astype(int)
```

# Rating

```python
avg_rating=data.rating.mean()
```

```python
data.rating = data.rating.replace(-1.0, np.nan)
data.rating=data.rating.fillna(avg_rating)
data.rating=data.rating.round(decimals=1)
data.rating
```

# Location

```python
data['location']=data. location.str.replace(" Australia Aus"," Australia")
data['location']=data. location.str.replace("India,In","India")
data['location']=data. location.str.replace("India In","India")
data['location']=data. location.str.replace("New York,Ny","New York")
data.location
```

# Established

```python
data.established=data.established.astype(str)
data.established = data.established.str.replace('-1','Unknown')
data.established
```

# Easy Apply

```python
data.easy_apply=data.easy_apply.str.replace("-1","FALSE")
data.easy_apply
```
